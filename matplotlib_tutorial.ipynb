## Using Figure Objects
- Figure is a top level class for all plot elements
- Figure class: https://www.tutorialspoint.com/matplotlib/matplotlib_figure_class.htm
- Axes class: https://www.tutorialspoint.com/matplotlib/matplotlib_axes_class.htm#:~:text=Advertisements,only%20be%20in%20one%20Figure.

```python
fig_1 = plt.figure(figsize=(5,4), dpi=100)

# left, bottom, width and height of the axes
axes_1 = fig_1.add_axes([0,0,0.9,0.9])
axes_1.set_xlabel('days')
axes_1.set_ylabel('days squared')
axes_1.set_title('days squared chart')
axes_1.plot(x_1, y_1, label='x/x^2')
axes_1.plot(y_1, x_1, label='x^2/x')

# upper right: 1, upper_left: 2 , lower left: 3, lower right: 4
axes_1.legend(loc=0)
```

## Pie Chart

```python
import random
fig_6 = plt.figure(figsize=(8,5), dpi=100)
axes_6 = fig_6.add_axes([0.1,0.1,0.9,0.9])

# create a pie chart of the number of Pokemon
types = ['Water', 'Normal', 'Flying', 'Grass', 'Psychic', 'Bug', 'Fire', 'Poison', 'Ground', 'Rock', 'Fighting', 'Dark', 'Steel', 'Electric', 'Dragon', 'Fairy', 'Ghost', 'Ice']

poke_num = [133, 109,101,98,85,77,68,66,65,60,57,54,53,51,50,50,46,40]
colors = []
for i in range(18):
    rgb = (random.uniform(0,0.5), random.uniform(0,0.5), random.uniform(0,0.5))
    colors.append(rgb)
    
explode = [0] * 18
explode[0] = 0.2

wedges, texts, autotexts = plt.pie(poke_num,explode=explode, labels=types, colors=colors, autopct='%1.0f%%', shadow=True, startangle=140, textprops=dict(color='w'))
plt.legend(wedges, types, loc='right', bbox_to_anchor=(1,0, 0.5, 1))
```

## 3D Surface

### 3D Scatter plot

```python
fig_9 = plt.figure(figsize=(8,5), dpi=100)
axes_9 = fig_9.add_axes([0.1,0.1,0.9,0.9], projection='3d')

z_3 = 15*np.random.random(100)
x_3 = np.sin(z_3)*np.random.random(100)
y_3 = np.cos(z_3)*np.random.random(100)

axes_9.scatter3D(x_3, y_3, z_3, c=z_3, cmap='Blues')
```

### 3D Conture plot

```python
def get_z(x,y):
    return np.sin(np.sqrt(x**2+y**2))

x_4 = np.linspace(-6,6,30)
y_4 = np.linspace(-6,6,30)

x_4, y_4 = np.meshgrid(x_4, y_4)
z_4 = get_z(x_4, y_4)
```

```python
fig_10 = plt.figure(figsize=(8,5), dpi=100)
axes_10 = fig_10.add_axes([0.1,0.1,0.9,0.9], projection='3d')

axes_10.view_init(45,55)
axes_10.contour3D(x_4, y_4, z_4, 100, cmap='Blues')
# plt.show()
```

```python
fig_11 = plt.figure(figsize=(8,5), dpi=100)
axes_11 = fig_11.add_axes([0.1,0.1,0.9,0.9], projection='3d')

axes_11.view_init(45,55)

axes_11.set_xlabel('x')
axes_11.set_ylabel('y')
axes_11.set_zlabel('z')

axes_11.plot_wireframe(x_4, y_4, z_4, color='blue')
```

```python
fig_12 = plt.figure(figsize=(8,5), dpi=100)
axes_12 = fig_12.add_axes([0.1,0.1,0.9,0.9], projection='3d')

axes_12.view_init(45,55)

axes_12.set_xlabel('x')
axes_12.set_ylabel('y')
axes_12.set_zlabel('z')

axes_12.plot_surface(x_4, y_4, z_4, rstride=1, cstride=1, cmap='Blues')
```
